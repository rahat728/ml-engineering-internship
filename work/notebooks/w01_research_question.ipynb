## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import os

# Locate and read dataset safely across local or colab setups
data_path = "data/raw/content_refresh_anonymized.csv"
if not os.path.exists(data_path) and os.path.exists("../" + data_path):
    data_path = "../" + data_path

df = pd.read_csv(data_path)
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

# Extract foundational validation baselines
total_pages = len(df)
portfolio_decline_rate = df["is_declining_label"].mean() * 100

# Calculate high-exposure risk surface area (Stale for >= 6 months, yet highly visible)
stale_visible_mask = (df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)
risk_surface_pct = (df[stale_visible_mask].shape[0] / total_pages) * 100

print(f"Metric 1: Total Portfolio Size = {total_pages:,} items")
print(f"Metric 2: Baseline Portfolio Decline Rate = {portfolio_decline_rate:.2f}%")
print(f"Metric 3: High-Exposure Risk Surface (Stale + Visible) = {risk_surface_pct:.2f}% of inventory")
```
